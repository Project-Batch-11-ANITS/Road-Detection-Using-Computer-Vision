```
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from google.colab.patches import cv2_imshow as imageShow
```

```
roadImage = cv.imread("/content/roadImage.jpg",cv.IMREAD_GRAYSCALE)
roadImage_GaussianBlured = roadImage.copy()
```

```
imageShow(roadImage)
```

```
roadImage.shape
```

```
height = roadImage.shape[0]
width = roadImage.shape[1]
```

```
gauss_matrix_not_normalized = np.array([[0,1,2,1,0],[1,3,5,3,1],[2,5,9,5,2],[1,3,5,3,1],[0,1,2,1,0]])
```

```
gauss_matrix_not_normalized
```

```
gauss_matrix_normalized = (1.0/57)*(gauss_matrix_not_normalized)
```

```
gauss_matrix_normalized
```

```
# Ensuring Gauss Matrix as normalized
print(sum(gauss_matrix_normalized))
print(sum(sum(gauss_matrix_normalized)))
```

```
for i in np.arange(2,height-2):
  for j in np.arange(2,width-2):
    sum=0
    for k in np.arange(-2,3):
      for l in np.arange(-2,3):
        a = roadImage.item(i+k,j+l)
        p = gauss_matrix_normalized[2+k,2+l]
        sum+=(p*a)
      b = sum
      roadImage_GaussianBlured.itemset((i,j),b)
```

```
imageShow(roadImage_GaussianBlured)
```
